```python
import matplotlib.pyplot as plt

x=["Python", "C", "C++", "Java"]
y=[90,60,65,85]

c = ["r", "g", "b", "y"]

plt.xlabel("Language", fontsize=15)
plt.ylabel("Percentage", fontsize=15)
plt.title("Programmig Languages Popularity")

plt.bar(x,y, color=c, width = 0.4, edgecolor="black", alpha = 0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["Python", "C", "C++", "Java"]
y=[90,60,65,85]
z=[20,40,60,70]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Language", fontsize=15)
plt.ylabel("Percentage", fontsize=15)
plt.title("Programmig Languages Popularity")

plt.bar(p,y, width, color="y", label="Popularity")
plt.bar(p1,z, width, color="r", label="Popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend(loc="upper center")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "C", "C++", "Java"]
y = [90, 60, 65, 85]
z = [20, 40, 60, 70]

height = 0.2
p = np.arange(len(x))
p1 = [j + height for j in p]

plt.ylabel("Language", fontsize=15)
plt.xlabel("Percentage", fontsize=15)
plt.title("Programming Languages Popularity")

plt.barh(p, y, height, color="y", label="Popularity")
plt.barh(p1, z, height, color="r", label="Popularity1")

plt.yticks(p + height / 2, x, rotation=20)
plt.legend(loc="center right")
plt.tight_layout()
plt.show()
```
