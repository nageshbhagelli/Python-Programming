```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x, tick_labels=["python"])
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x, vert=False)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x, notch=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x, tick_labels=["python"], widths=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, tick_labels=["python"], patch_artist=True, showmeans=True, sym="g")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x, tick_labels=["python"], showmeans=True, boxprops=dict(color="r"), capprops=dict(color="g"), whiskerprops=dict(color="g"))
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]
x1 = [10,30,45,68,79,55,23]
y = [x,x1]

plt.boxplot(y, tick_labels=["python", "java"])
plt.show()
```
