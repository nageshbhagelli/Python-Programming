```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8]
y = [2,3,1,4,5,3,6,1]
c = ["r","y","g","b","m","r","y","g"]
size = [100,250,400,200,300,500,400,150]
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Scatter Plot")
plt.scatter(x,y, color=c, s=size, alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8]
y = [2,3,1,4,5,3,6,1]
c = [35,23,44,76,18,96,66,10]
size = [100,250,400,200,300,500,400,150]

plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Scatter Plot")

plt.scatter(x,y, c=c, s=size, cmap="BrBG")
t=plt.colorbar()
t.set_label("Color Bar", fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8]
y = [2,3,1,4,5,3,6,1]
z = [5,3,7,1,4,2,9,6]
size = [100,250,400,200,300,500,400,150]

plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Scatter Plot")

plt.scatter(x,y, c="r", label="Apple")
plt.scatter(x,z, c="g", label="Mango")
plt.legend(loc="upper left")
plt.show()
```
