```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
area1 = [2,3,7,4,5,1]
area2 = [3,2,5,6,8,2]
area3 = [1,5,7,8,3,4]

l = ["area1","area2","area3"]

plt.stackplot(x,area1,area2,area3, labels=l)
plt.title("Stack/Area Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```
