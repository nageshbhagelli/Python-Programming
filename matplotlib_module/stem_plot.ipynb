```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8]
y = [2,3,5,6,4,3,2,1]

plt.stem(x, y, label="python")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8]
y = [2,3,5,6,4,3,2,1]

plt.stem(x, y, label="python", linefmt=":", markerfmt="r*", basefmt="g", bottom=-5)
plt.legend()
plt.show()
```
