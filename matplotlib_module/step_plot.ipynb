```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [1,2,0,4,5,3]

plt.title("Step Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.step(x,y, color="r", marker="o", ms=10, mfc="g", label="python")
plt.legend(loc=2)
plt.grid()
plt.show()
```
